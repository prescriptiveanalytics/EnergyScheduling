```python
import pandas as pd
import pickle
```

```python
input_file_name = "data/cluster_0_per_hour_n=5.pkl"
with open(input_file_name, "rb") as inf:
    df = pickle.load(inf)
    df.head()
```

```python
df.head()
```

```python
grouped = df[['Hour', 'kwh']].groupby(by='Hour').median()
values = grouped['kwh'].values
consumption = {}
for k, v in enumerate(values):
    print(v)
    consumption[k] = {}
    for m in range(0, 60, 15):
        consumption[k][m] = v/4

consumption
```
